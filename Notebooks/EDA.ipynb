```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('C:\\xampp\\htdocs\\git\\MobilePhonePricePrediction\\Data\\Clean\\cleaned_cellphone.csv')
df.head()
```

```python
# Display information about the DataFrame
df.info()
df.describe()
df.head()
```

```python
# Plotting the distribution of the 'price' column
plt.figure(figsize=(8,5))
sns.histplot(df['price'], kde=True)
plt.title('Handphone Price Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting the distribution of all numeric features
df.hist(figsize=(15,10), bins=20)
plt.suptitle("Numeric Features Distribution")
plt.show()
```

```python
# Plotting the correlation heatmap of numeric features
plt.figure(figsize=(12,10))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

```python
# Plotting handphone price range distribution
plt.figure(figsize=(8,5))
sns.boxplot(x='price_range', y='price', data=df)
plt.title('Price range distribution')
plt.show()
```

```python
# Scatter plots for handphone specifications vs price
features = ['ram', 'battery', 'rear_cam', 'cpu_freq', 'resolution', 'internal_mem', 'cpu_core', 'front_cam']
plt.figure(figsize=(16,10))
for i, col in enumerate(features):
    plt.subplot(2, 4, i+1)
    sns.scatterplot(x=col, y='price', data=df)
    plt.title(f'{col.capitalize()} vs Price')

plt.tight_layout()
plt.show()
```

```python
# Pairplot for selected features and price
sns.pairplot(df, vars=features + ['price'])
plt.show()
```

```python
# Outlier check
for col in ['battery', 'ram']:
    plt.subplot(2, 1, 1 if col == 'battery' else 2)
    sns.boxplot(x=df[col])
    plt.title(f'Outlier Check: {col}')
    plt.show()
```
